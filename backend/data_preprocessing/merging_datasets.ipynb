```python
import pandas as pd
crazy = crazy.rename(columns={
    "Model":"Title"
})
fusion = fusion.rename(columns={
    "Fuel": "Fuel Type",
    "Image_URL": "Image",
    "URL": "Link",

})

olx1 = olx1.rename(columns={
    'Owner':'Ownership'
})

# Step 2: Define master schema
all_columns = [
    'Title', 'Link', 'Location', 'Price', 'Image', 'Fuel Type', 'Type',
    'Year', 'Kms Covered', 'Brand', 'Exterior Color',
    'Interior Color', 'Registration', 'Ownership'
]

# Step 3: Reindex each dataset
olx1 = olx1.reindex(columns=all_columns)
olx2 = olx2.reindex(columns=all_columns)
fusion = fusion.reindex(columns=all_columns)
crazy = crazy.reindex(columns=all_columns)

# Step 4: Concatenate
merged_df = pd.concat([olx1, olx2, fusion, crazy], ignore_index=True)

# Step 5: Add Indian-style price display
def indian_format(num):
    try:
        s = str(int(num))
    except:
        return None
    r = s[-3:]
    l = s[:-3]
    while len(l) > 2:
        r = l[-2:] + "," + r
        l = l[:-2]
    if l:
        r = l + "," + r
    return "₹" + r

merged_df["price_display"] = merged_df["Price"].apply(indian_format)

# Save or inspect
merged_df.to_csv("merged_dataset.csv", index=False)
print(merged_df.head())
```

## Removing duplicates from merged datasets

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data/merged_datasets/merged_luxe_dataset.csv',index_col=0)
```

```python
df.head()
```

```python
df.shape
```

```python
df.head()
```

```python
from urllib.parse import urlparse

df['Platform'] = df['Link'].apply(lambda x: urlparse(x).netloc.replace('www.', ''))

olx_df = df[df['Platform'].str.contains('olx', case=False, na=False)]

olx_df = olx_df.drop_duplicates(subset=['Image','Year','Title','Price','Kms Covered'])

non_olx_df = df[~df['Platform'].str.contains('olx', case=False, na=False)]

cleaned_df = pd.concat([non_olx_df, olx_df], ignore_index=True)

print(cleaned_df)
```

```python
cleaned_df.to_csv('../../data/merged_datasets/merged_luxe_dataset.csv')
```
